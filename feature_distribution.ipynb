### First a simple histogram of the musicians data from spotify

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(5,5))
fig = sns.histplot(   
        data=mus_nodes,
        x='popularity',
    )
fig.set(
        xlabel='Popularity',
    )
fig.get_figure().savefig('pics/features/spotify_popularity.png',dpi=300)
```

### Then relation between popularity and average number of followers

```python
plt.figure(figsize=(5,5))
fig = sns.lineplot(
        data=mus_nodes,
        x='popularity',
        y='followers',
        ci='sd'
    )
fig.set(
        xlabel='Popularity',
        ylabel='Follower Count'
    )
            
fig.get_figure().savefig('pics/features/spotify_popularity_and_followers.png',dpi=300)
```
