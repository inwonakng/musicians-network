## Check basic distributions of our data

```python
import pandas as pd
import numpy as np
from utils import MusicDataLoader
from collections import Counter

loader = MusicDataLoader()
mus_nodes = loader.get_features(n_release_bins=100)
mus_nodes.network_rank = mus_nodes.network_rank.fillna(0)
```

### First a simple histogram of the musicians data from spotify

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(5,5))
fig = sns.histplot(   
        data=mus_nodes,
        x='popularity',
    )
fig.set(
        xlabel='Popularity',
    )
fig.get_figure().savefig('pics/features/spotify_popularity.png',dpi=300)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(5,5))
fig = sns.histplot(   
        data=mus_nodes,
        x='followers',
    )
fig.set(
        xlabel='Follower Count',
        xscale='log',
    )
fig.get_figure().savefig('pics/features/spotify_followers.png',dpi=300)
```

### Then relation between popularity and average number of followers

```python
plt.figure(figsize=(5,5))
fig = sns.lineplot(
        data=mus_nodes,
        x='popularity',
        y='followers',
        ci='sd'
    )
fig.set(
        xlabel='Popularity',
        ylabel='Follower Count'
    )
            
fig.get_figure().savefig('pics/features/spotify_popularity_and_followers.png',dpi=300)
```
