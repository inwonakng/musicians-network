```python
import pandas as pd
import pandas as pd
from tqdm import tqdm
from utils import MusicDataLoader

loader = MusicDataLoader()
mus_nodes = loader.get_features(n_release_bins=100)
mus_nodes.network_rank = mus_nodes.network_rank.fillna(0)

billboard = pd.read_csv('data/charts.csv')
billboard.date = pd.to_datetime(billboard.date)
billboard['name'] = billboard.artist.str.split('&',n=1).str[0].str.split('Featuring').str[0].str.strip()
billboard = billboard.merge(mus_nodes,on='name')
after_nov = billboard[billboard.date > '2021-11-01']
mus_nodes['is_billboard'] = mus_nodes.id.isin(after_nov.id)
mus_nodes.is_billboard.sum()
```

```python
import networkx as nx

G = nx.read_gml(f'data/musician-graph/graph_100.gml')

pops = nx.get_node_attributes(G,'popularity')
fols = nx.get_node_attributes(G,'followers')
```

```python
import numpy as np
maxfol = np.array(list(fols.values())).max()
```

```python
len(list(G.nodes)), len(fols), len(pops)
```

```python
node = 'Drake'

df = pd.DataFrame()

vals = []

for node in G.nodes:
    neigh_pops = []
    neigh_fols = []
    neigh_numrel = []

    for neigh in G.neighbors(node):
        if neigh in pops:
            neigh_pops.append(pops[neigh])
        if neigh in fols:
            neigh_fols.append(fols[neigh])
        if neigh in mus_nodes.name.values:
            neigh_numrel.append(mus_nodes[mus_nodes.name == neigh].num_release.values[0])
            
    neigh_pops = np.array(neigh_pops)/100
    neigh_fols = np.array(neigh_fols)/maxfol
    
    if len(neigh_pops) == 0:
        neigh_pops = np.array([0])
        
    if len(neigh_fols) == 0:
        neigh_fols = np.array([0])

    vals.append([
        neigh_pops.mean(), 
        neigh_pops.std(), 
        neigh_pops.var(),
        neigh_fols.mean(), 
        neigh_fols.std(), 
        neigh_fols.var()
    ])
```
